```python
df = pd.read_csv("../datasets/movies_metadata.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
m = df["vote_count"].quantile(0.98)
```

```python
m
```

```python
reasonably_long = df[(df["runtime"] > 45) & (df["runtime"] < 300)]
```

```python
reasonably_long.shape
```

```python
reasonably_voted = df[df["vote_count"] > m]
```

```python
reasonably_voted.shape
```

```python
avg = reasonably_voted["vote_average"].mean()
```

```python
avg
```

```python
def weighted_rating(x, m=m, avg=avg):
    v = x["vote_count"]
    R = x["vote_average"]
    return (v/(v+m) * R) + (m/(m+v) * avg)
```

```python
reasonable = reasonably_voted.copy()
```

```python
reasonable['score'] = reasonable.apply(weighted_rating, axis=1)
```

```python
reasonable.shape
```

```python
reasonable.sort_values("score", ascending=False)
```

```python
reasonable.head(250)
```

```python
reasonable = reasonable.set_index("title")
```

```python
reasonable = reasonable.sort_values("score", ascending=False)
```
